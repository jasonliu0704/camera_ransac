# 3(a): use SIFT to detect and draw matches

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline


img1 = cv2.imread('road1.png',0)          # queryImage
img2 = cv2.imread('road2.png', 0) # trainImage

# Initiate SIFT detector
orb = cv2.ORB_create()

# find the keypoints and descriptors with SIFT
kp1, des1 = orb.detectAndCompute(img1,None)
kp2, des2 = orb.detectAndCompute(img2,None)
```

```python
# BFMatcher with default params
bf = cv2.BFMatcher()
matches = bf.knnMatch(des1,des2, k=2)

# Apply ratio test
good = []
for m,n in matches:
    if m.distance < 0.75*n.distance:
        good.append([m])

# cv2.drawMatchesKnn expects list of lists as matches.
img3 = cv2.drawMatchesKnn(img1,kp1,img2,kp2,good,outImg=img3,flags=2)

plt.imshow(img3),plt.show()
cv2.imwrite('imageMatches.jpg',img3)
```

# 3(b): get fundamental matrix from SIFT and draw epipolar line

```python
# get point of interets from amtches
pts1 = []
pts2 = []

for m in good:
    pts1.append(kp1[m[0].queryIdx].pt)
    pts2.append(kp2[m[0].trainIdx].pt)
pts1 = np.int32(pts1)
pts2 = np.int32(pts2)
F, mask = cv2.findFundamentalMat(pts1,pts2,cv2.FM_LMEDS)
```

```python
def drawlines(img1,img2,lines,pts1,pts2):
    ''' img1 - image on which we draw the epilines for the points in img2
        lines - corresponding epilines '''
    r,c = img1.shape
    img1 = cv2.cvtColor(img1,cv2.COLOR_GRAY2BGR)
    img2 = cv2.cvtColor(img2,cv2.COLOR_GRAY2BGR)
    for r,pt1,pt2 in zip(lines,pts1,pts2):
        color = tuple(np.random.randint(0,255,3).tolist())
        x0,y0 = map(int, [0, -r[2]/r[1] ])
        x1,y1 = map(int, [c, -(r[2]+r[0]*c)/r[1] ])
        img1 = cv2.line(img1, (x0,y0), (x1,y1), color,1)
        img1 = cv2.circle(img1,tuple(pt1),5,color,-1)
        img2 = cv2.circle(img2,tuple(pt2),5,color,-1)
    return img1,img2
```

```python
# Find epilines corresponding to points in right image (second image) and
# drawing its lines on left image
lines1 = cv2.computeCorrespondEpilines(pts2.reshape(-1,1,2)[0:4], 2,F)
lines1 = lines1.reshape(-1,3)
img5,img6 = drawlines(img1,img2,lines1,pts1,pts2)
# Find epilines corresponding to points in left image (first image) and
# drawing its lines on right image
lines2 = cv2.computeCorrespondEpilines(pts1.reshape(-1,1,2)[0:4], 1,F)
lines2 = lines2.reshape(-1,3)
img3,img4 = drawlines(img2,img1,lines2,pts2,pts1)
plt.subplot(121),plt.imshow(img5)
plt.subplot(122),plt.imshow(img3)
plt.show()
plt.imsave("3(b)1.png", img5)
plt.imsave("3(b)2.png", img3)
```

# 3(c): run RANSAC and draw match line and epipolar line

```python
import random

def ransac():
    threshold = 0.001
    N = 1000
    inlier_g = 0
    for i in range(0,N):
        # get 8 samples
        rand_matches = random.sample(good, 8)
        p = []
        q = []
        for m in rand_matches:
            p.append(kp1[m[0].queryIdx].pt)
            q.append(kp2[m[0].trainIdx].pt)
        p = np.int32(p)
        q = np.int32(q)
        # calculate F
        A = []
        for k in range(0,8):
            u = p[k][0]
            v = p[k][1]
            up = q[k][0]
            vp = q[k][1]
            A.append([u*up, v*up, up, u*vp, v*vp, vp, u, v, 1])
        U, s, V = np.linalg.svd(A, full_matrices=True)
        F = np.transpose(V[:, 8].reshape((3, 3)))
        U, s, V = np.linalg.svd(F, full_matrices=True)
        F = U.dot(np.diag([s[0], s[1], 0])).dot(np.transpose(V))

        # inliner counting
        inlier_count = 0
        for k in range(0, len(pts1)):
            x1 = np.array([pts1[k][0], pts1[k][1], 1])
            x2 = np.array([pts2[k][0], pts2[k][1], 1])
            if(x1.dot(F).dot(np.transpose(x2)) < threshold):
                inlier_count +=1
        # compare previous model
        if(inlier_count > inlier_g):
            inlier_g = inlier_count
            F_g = F
    return F_g
```

```python
# draw inlier match
threshold = 0.001
Fundamental = ransac()
new_matches = []
new_pts1 = []
new_pts2 = []
for m in good:
    pt1 = kp1[m[0].queryIdx].pt
    pt2 = kp2[m[0].trainIdx].pt
    x1 = np.array([pt1[0], pt1[1], 1])
    x2 = np.array([pt2[0], pt2[1], 1])
    if(x1.dot(Fundamental).dot(np.transpose(x2)) < threshold):
        new_matches.append(m)
        new_pts1.append(pt1)
        new_pts2.append(pt2)
    
img4 = 0
img4 = cv2.drawMatchesKnn(img1,kp1,img2,kp2,new_matches, outImg = img4)
plt.imshow(img4),plt.show()
cv2.imwrite('imageMatches_3(c).jpg',img4)
```

```python
#draw epipolar line
#randomly select 5 points
pts1_rand = np.int32(new_pts1[0:5])
pts2_rand = np.int32(new_pts2[0:5])

lines1 = cv2.computeCorrespondEpilines(pts2_rand.reshape(-1,1,2), 1,F)
lines1 = lines1.reshape(-1,3)
img5,img6 = drawlines(img1,img2,lines1,pts1_rand,pts2_rand)
# Find epilines corresponding to points in left image (first image) and
# drawing its lines on right image
lines2 = cv2.computeCorrespondEpilines(pts1_rand.reshape(-1,1,2), 2,F)
lines2 = lines2.reshape(-1,3)
img3,img4 = drawlines(img2,img1,lines2,pts2_rand,pts1_rand)
plt.subplot(121),plt.imshow(img5)
plt.subplot(122),plt.imshow(img3)
plt.show()
plt.imsave("3(c)epipolar1.png", img5)
plt.imsave("3(c)epipolar2.png", img3)
```
